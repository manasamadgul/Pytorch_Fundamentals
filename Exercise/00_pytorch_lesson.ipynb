```python
#Matrix Mulitplication
 # 1.way 1 : element wise multiplication (each element by a scalar))
 # 2.way 2 : Matrix multiplication/dot product
```

```python
#way 1 : element wise
tensor = torch.tensor([1,2,3])
tensor*tensor
```

```python
#way2 dot product or matrix multiplication
%%time
torch.matmul(tensor,tensor)
```

```python
tensor1 = torch.rand((3,3), dtype=torch.float16)
print(tensor1)
print(tensor1.ndim)
tensor2 = torch.rand((3,2), dtype= torch.float16)
print(tensor2)
print(tensor2.ndim)
torch.matmul(tensor1,tensor2).shape
```

```python
#transpose can be used to change the shape of a matrix. this can be very useful when we are doing matrix multiplication and size errors are faced
#matrix multiplication can also be written as matmul,mm(short form) or @
#.T --> transpose operation
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_B = torch.tensor([[7,10],[8,11],[9,12]])
torch.matmul(tensor_A,tensor_B.T)
```

```python
# Create a tensor
import torch
x = torch.arange(1., 8.)
x, x.shape
```

```python
# Change view (keeps same data as original but changes view)
# See more: https://stackoverflow.com/a/54507446/7900723
z = x.view(1, 7)
z, z.shape
```

```python
z[:,0] = 5
z,x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0) # try changing dim to dim=1 and see what happens
x_stacked
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1) # try changing dim to dim=1 and see what happens
x_stacked
```
